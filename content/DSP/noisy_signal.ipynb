# Noisy Signals

This notebook explores properties of signals and noise, and reviews some basic DSP concepts.

```python
%pip install -q nbformat plotly

import numpy as np
from scipy.fft import fft, ifft, fftshift, fftfreq
import plotly.graph_objects as go
```

```python
# generate a random noise sequence
n_t = np.random.randn(10240)

# obtain its frequency spectrum through FFT
n_f = fft(n_t)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(len(n_t)), 
                         y=20*np.log10(np.abs(n_f))),)
fig.update_layout(yaxis_title='Amplitude (dB)')
```
